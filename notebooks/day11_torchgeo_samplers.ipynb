```python
from torchgeo.samplers import RandomGeoSampler

print(RandomGeoSampler)
```

```python
from torchgeo.samplers import GridGeoSampler

print(GridGeoSampler)
```

```python
patch_sizes = [128, 256, 512]

for size in patch_sizes:
    print(f"Patch Size: {size}x{size}")
```

```python
training_workflow = """
Sentinel-2
     ↓
RasterDataset
     ↓
RandomGeoSampler
     ↓
Patch Extraction
     ↓
Model Training
"""

print(training_workflow)
```

```python
inference_workflow = """
Sentinel-2
     ↓
RasterDataset
     ↓
GridGeoSampler
     ↓
Full Coverage
     ↓
Prediction
     ↓
Change Map
"""

print(inference_workflow)
```

```python
comparison = {
    "RandomGeoSampler": "Training",
    "GridGeoSampler": "Inference"
}

for sampler, use in comparison.items():
    print(f"{sampler} -> {use}")
```
